## Question 1


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import gamma
from scipy.stats import chi2
```

```python
def a1():
    n = 15
    p = 0.2
    k = np.arange(0, n + 1)
    pmf_values = binom.pmf(k, n, p)

    plt.figure(figsize=(8, 5))
    plt.stem(k, pmf_values, basefmt=" ")
    plt.title("PMF of the Binomial Distribution (n=15, p=0.2)", fontsize=14)
    plt.xlabel("Number of Successful Responses (k)", fontsize=12)
    plt.ylabel("Probability", fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
```

```python
a1()
```

```python
def b1():
  n = 15
  p = 0.2
  k = np.arange(0,n+1)
# Range of probabilities
  probabilities = np.arange(0.1, 1.0, 0.1)

  # Plotting the PMF for different probabilities
  plt.figure(figsize=(12, 8))

  for p in probabilities:
      pmf_values = binom.pmf(k, n, p)
      plt.plot(k, pmf_values, label=f'p={p:.1f}', marker='o')

  plt.title("PMF for Different Success Probabilities", fontsize=14)
  plt.xlabel("Number of Successful Responses (k)", fontsize=12)
  plt.ylabel("Probability", fontsize=12)
  plt.legend(title="Probability of Success (p)", fontsize=10)
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.show()
```

```python
b1()
```

```python
def c1():
  # Parameters
  p = 0.05
  n_values = [10, 20, 50, 200]

  plt.figure(figsize=(12, 8))

  for n in n_values:
      # Simulate success rates
      k = np.arange(0, n + 1)
      pmf_values = binom.pmf(k, n, p)
      success_rate = k / n
      plt.plot(success_rate, pmf_values, label=f'n={n}', marker='o')

  plt.title("Success Rate Distribution for Different n", fontsize=14)
  plt.xlabel("Success Rate (Y = X/n)", fontsize=12)
  plt.ylabel("Probability", fontsize=12)
  plt.legend(title="Number of Customers Contacted (n)", fontsize=10)
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.show()
```

```python
c1()
```

```python
def a2():
  # Parameters
  alpha = 5  # Shape parameter
  theta = 4  # Scale parameter

  # Calculations
  mean = alpha * theta
  variance = alpha * theta**2
  std_dev = np.sqrt(variance)

  # Display results
  print(f"Mean (Expected Lifetime): {mean}")
  print(f"Variance: {variance}")
  print(f"Standard Deviation: {std_dev}")
```

```python
def c2():
  alpha = 5
  theta = 4
  x = np.linspace(0.1, 50, 500)

  # PDF calculation
  pdf_values = gamma.pdf(x, alpha, scale=theta)

  # Plotting the PDF
  plt.figure(figsize=(10, 6))
  plt.plot(x, pdf_values, label=f'α={alpha}, θ={theta}', color='b')
  plt.title("PDF of Gamma Distribution", fontsize=14)
  plt.xlabel("Battery Lifetime (hours)", fontsize=12)
  plt.ylabel("Density", fontsize=12)
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.legend()
  plt.show()
```

```python
c2()
```

```python
def d2():
  x = np.linspace(0.1, 50, 500)
  # Different parameter sets
  params = [
      (10, 0.9),
      (7, 2),
      (1.5, 0.2)
  ]

  plt.figure(figsize=(10, 6))

  for alpha, theta in params:
      # PDF calculation
      pdf_values = gamma.pdf(x, alpha, scale=theta)
      plt.plot(x, pdf_values, label=f'α={alpha}, θ={theta}')

  plt.title("PDF of Gamma Distribution with Varying Parameters", fontsize=14)
  plt.xlabel("Battery Lifetime (hours)", fontsize=12)
  plt.ylabel("Density", fontsize=12)
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.legend()
  plt.show()
```

```python
d2()
```

## Question 3

```python
def a3():
  # Degrees of freedom
  df = 10

  # Calculations
  mean = df
  variance = 2 * df
  std_dev = np.sqrt(variance)

  # Percentiles
  median = chi2.ppf(0.5, df)
  q1 = chi2.ppf(0.25, df)
  q3 = chi2.ppf(0.75, df)

  # Display results
  print(f"Mean: {mean}")
  print(f"Variance: {variance}")
  print(f"Standard Deviation: {std_dev}")
  print(f"Median: {median}")
  print(f"First Quartile (Q1): {q1}")
  print(f"Third Quartile (Q3): {q3}")
```

```python
a3()
```

```python
def b3():
    # Range of x values
  x = np.linspace(0, 24, 500)
  df = 10
  mean = df
  median = chi2.ppf(0.5,df)
  q1 = chi2.ppf(0.25, df)
  q3 = chi2.ppf(0.75, df)

  # PDF calculation
  pdf_values = chi2.pdf(x, df)

  # Plotting the PDF
  plt.figure(figsize=(10, 6))
  plt.plot(x, pdf_values, label=f'χ² (df={df})', color='b')

  # Highlight mean, median, Q1, and Q3
  plt.axvline(mean, color='r', linestyle='--', label=f'Mean = {mean}')
  plt.axvline(median, color='g', linestyle='--', label=f'Median = {median:.2f}')
  plt.axvline(q1, color='orange', linestyle='--', label=f'Q1 = {q1:.2f}')
  plt.axvline(q3, color='purple', linestyle='--', label=f'Q3 = {q3:.2f}')

  plt.title("PDF of Chi-Squared Distribution", fontsize=14)
  plt.xlabel("Battery Lifetime (hours)", fontsize=12)
  plt.ylabel("Density", fontsize=12)
  plt.legend()
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.show()
```

```python
b3()
```
